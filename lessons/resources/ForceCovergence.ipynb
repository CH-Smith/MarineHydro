## Pressure force estimates

We ran into D'Alembert's paradox earlier, which states that there is zero drag force in potential flow, regardless of the body shape. However, in a real flow we have a boundary layer. This induces friction force, which can be significant for streamlined bodies. However, the real impact is the separation, after which the potential flow solution is completely invalid. 

---
![Prandtl circle](resources/prandtl.png)

---

A function can be defined using the routines from the previous notebooks to identify the attached flow region from the detached flow region.

```python
def separation_index(panels): # symmetric flow only
    top,bottom = split(panels)
    delta,lam,iSep = panel_march(top)
    return len(top)-iSep-1
```

```python
circle = make_circle(64)
solve_gamma_kutta(circle)
plot_flow(circle)

iSep = separation_index(circle)
for i in range(iSep,len(circle)-iSep):  # attached flow region
    pyplot.scatter(circle[i].xc,circle[i].yc,s=100,c='g')
for i in range(-iSep,iSep):             # separated region
    pyplot.scatter(circle[i].xc,circle[i].yc,s=100,c='r')
```

Recall that the pressure drag coefficient depends on $c_p$. Using the vortex panel method we can determine the potential flow solution for $c_p$, but what does $c_p$ look like in a real flow? 

---
![Pressure coefficient](resources/graphics5.png)

---
I've sketched the results for the flow around a circular cylinder above. The measured pressure at the front of a body in a viscous fluid is fairly well predicted by potential flow. 

However, the pressure coefficient completely deviates from the potential flow prediction near the point of separation. **Indeed it remains essentially constant in the separated flow region.**

The simplest way to make nonzero pressure force estimates is to assume the pressure in the separated region is constant.

 - **Modify** the $C_D$ function from the Pressure notebook to generate non-zero estimates.

##### Solution #1

```python
def c_p(gamma): return 1-gamma**2

def c_p_sep(gamma,iSep):
    c_p_sep = c_p(gamma)
    for i in range(-iSep,iSep): 
        c_p_sep[i] = c_p_sep[iSep]
    return c_p_sep

def C_D( panels ):
    ds = numpy.gradient(distance(panels))
    gamma = numpy.array([p.gamma for p in panels])
    sy = [p.sy for p in panels]
    yc = [p.yc for p in panels]
    t = max(yc)-min(yc)        
    iSep = separation_index(panels)
    C_D = sum(c_p_sep(gamma,iSep)*sy*ds)/t
    return C_D
```

```python
for t_c in [8,4,2,1,0.5,0.25,0.125]:
    ellipse = make_circle(64,t_c=t_c)
    solve_gamma_kutta(ellipse)
    print('ellipse C_D = '+'%.2f'%(C_D(ellipse)))
```

## Friction drag

The friction drag routine from the boundary layer notebook also needs to be adjusted. First, on a curved boundary layer, the wall stress points in the tangent direction, not the x-direction. Second, the stress is only integrated up to the point of separation. Therefore the friction drag force is:

$$ F_f = \int_{\Omega_a} \tau_w s_x ds $$ 

where $\Omega_a$ is the region of attached flow. The friction drag coefficient is then

$$ C_F = \frac {F_f}{\frac 12 \rho U^2_\infty S} $$

where $S$ is the circumference (2D surface area) of the body. 

##### Your turn #2

- **Complete** the $C_F$ function below 
- 

```python
from BoundaryLayer import march,df_0

def tau_w(u_e,delta,lam,nu):  # really tau_w/rho
    tau_w = u_e*nu/delta*df_0(lam)
    return tau_w

def C_F( panels, Re ):
    top,bottom = split(panels)
    s = distance(top)
    ds = numpy.gradient(s)
    sx = [p.sx for p in top]
    u_e = numpy.array([-p.gamma for p in top])
    nu = s[-1]/Re
    
    delta,lam,iSep = march(s,u_e,nu)
    val = tau_w(u_e,delta,lam,nu)*sx*ds
    C_F = 2*sum(val[:iSep+1])/s[-1]
    return C_F
```

```python
print('C_F = '+'%.2e'%C_F(circle,1e5))
```

```python
for t_c in [8,4,2,1,0.5,0.25,0.125]:
    ellipse = make_circle(64,t_c=t_c)
    solve_gamma_kutta(ellipse)
    s = distance(ellipse)
    print('ellipse C_F = '+'%.2e'%(C_F(ellipse,1e5)))
print('plate C_F = '+'%.2e'%(1.33/numpy.sqrt(1e5)))
```
